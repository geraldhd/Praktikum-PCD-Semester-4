```python
import cv2 as cv 
import matplotlib.pyplot as plt 
import numpy as np 
```

Code di atas digunakan untuk

```python
image = plt.imread("marc.jpg") 
plt.subplot(121)
plt.title("Original image")
plt.imshow(image)
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY) 
plt.subplot(122)
plt.title("Gray image")
plt.imshow(gray, cmap="gray") 
```

Code di atas digunakan untuk:

Menggunakan fungsi imread dari plt untuk membaca citra yang akan diolah. kemudian disimpan di variabel image. Lalu menggunakan fungsi cvtColor dari cv untuk memuat gray scale dari citra dan menyimpannya di variabel gray. Dan terakhir menampilkan gambar gray dengan skala abu-abu

```python
sumgray = np.zeros(256) 
for i in range(256): 
    for j in range(256): 
        value = gray[i,j] 
        sumgray[value] += 1 
sumgray.sum() 
grayvalue = [i for i in range(256)] 
plt.bar(grayvalue, sumgray) 
```

Pada peulangan tersebut akan otomatis menambahkan nilai +1 jika terdapat value yang sama pada satu citra tersebut. ## Soal Nomor 2

 Lakukan ekualisasi histogram pada citra 1 dan tampilkan citra hasil ekualisasinya

```python
hist = np.zeros(256).astype(int) 
height, width = gray.shape 
#=================================================
```

Sebelum melakukan proses ekualisasi kita perlu membuat variabel "hist" yang akan menyimpan nilai array dengan panjang 256 dengan tipe data integer untuk semua data pada tiap indeks. lalu melakukan deklarasi variabel h dan w yang akan menyimpan nilai tinggi dan lebar pada citra dengang bantuan ".shape". 

```python
for i in range(height): 
    for j in range(width): 
        hist[gray[i,j]] +=1
#=================================================
```

Lalu kita perlu melakukan perulangan yang berfungsi untuk mengakses tiap indeks dari citra dengan menggunakan nested for dengan parameter variabel h dan w untuk mengakses baris dan kolom dalam citra.

```python
kumulatif = np.zeros(256).astype(int) 
for i in range(256): 
    kumulatif[i] = np.sum(hist[0:i+1]) 
#================================================
```

Lalu membuat variabel array untuk menyimpan kumulatif dari tiap indeks citra dengan tipe data integer. setelah itu kita perlu membuat kondisi perulangan untuk mengakses tiap indeks baris pada citra dan di dalam perulangan tersebut terdapat statement yang akan menjumlahkan setiap indeks dengan seluruh indeks sebelumnya

```python
goal= kumulatif*(256-1)/(height*width) 
goal= np.floor(goal).astype(int) 
```

Lalu kita perlu membuat variabel yang akan menyimpan hasil kali kumulasi dengan panjang indeksnya kemudian dibagi dengan ukuran dari citra (goal). setelah itu langsung membulatkan hasil tersebut ke bawah dan mengubah tipe datanya ke integer dengan fungsu "np.floor"

```python
result = np.zeros(gray.shape).astype(int) 
for i in range(height): 
    for j in range(width): 
        result[i,j] = goal[gray[i,j]]
plt.imshow(result, cmap="gray") 
```

sama seperti sebelumnya kita pelu membuat variabel array 0 dengan nama result dengan panjang indeks sama dengan citra abu-abu yang sedang diolah dan dengan tipe data integer. lalu kita perlu melakukan perulangan nested for dengan parameter h dan w untuk mengunjungi tiap indeks. di dalam perulangan tersebut akan menyimpan seluruh hasil kumulatif per indeks ke variabel result dan terakhir langsung tampilkan hasil ekualisasi dengan fungsi "plt.imshow(result, cmap="gray") "

```python
sumgray = np.zeros(256) 
for i in range(256): 
    for j in range(256): 
        value = result[i,j] 
        sumgray[value] += 1 
sumgray.sum() 
grayvalue = [i for i in range(256)] 
#===================================
grayscale= gray

derajat = np.zeros(256)

for i in range(256):
    derajat[i]=i

in_gray=np.zeros(256)
for row in grayscale:
    for col in row: 
        in_gray[col]+=1
#===================================
plt.subplot(121)
plt.xlabel("Sesudah Ekualisasi")
plt.ylabel("jumlah pixel")
plt.bar(grayvalue,sumgray) 
plt.subplot(122)
plt.xlabel("Sebelum Ekualisasi") 
plt.ylabel("jumlah pixel")
plt.bar(derajat,in_gray) 
```

## Soal Nomor 3

Lakukan spesifikasi/matching histogram pada citra 1 berdasarkan pada histogram citra 2. 

```python
haechan = plt.imread('haechan.jpg')
plt.subplot(121)
plt.title("Original image")
plt.imshow(haechan) 
gray_haechan = cv.cvtColor(haechan, cv.COLOR_RGB2GRAY) 
plt.subplot(122)
plt.title("Gray image")
plt.imshow(gray_haechan, cmap='gray') 
```

Code di atas digunakan untuk:

Menggunakan fungsi imread dari plt untuk membaca citra yang akan diolah. kemudian disimpan di variabel image2. Lalu menggunakan fungsi cvtColor dari cv untuk memuat gray scale dari citra dan menyimpannya di variabel gray. Dan terakhir menampilkan gambar gray dengan skala abu-abu

Sebelum melakukan perubahan kita perlu menampilkan gambar sebelum berubah sebagai perbandingan

```python
x, y = gray.shape
bit = 8
sumGray = np.zeros(2**bit)
for i in range(x): 
    for j in range(y): 
        grayValue = gray_haechan[i][j]
        sumGray[grayValue] += 1 
plt.bar([i for i in range(2**bit)],sumGray) 
plt.title("Histogram Citra 2")
plt.show()
```

Code di atas digunakan untuk:

Menghitung histogram dengan menggunakan perulangan nested for yang akan mengunjungi setiap pixel dari citra yang akan ditampilkan histogramnya.

```python
def find_goal(img, level):
    hist = np.zeros(level).astype(int)
    for i in range(x): 
        for j in range(y): 
            hist[img[i,j]] +=1
    goals = np.zeros(level).astype(int) 
    for i in range(level):
            goals[i] = np.sum( hist[0:i+1] )
    goal_value = goals*(level-1)/(x*y)
    goal_value = np.floor(goal_value).astype(int)
    return goal_value  
```

Code di atas digunakan untuk:

membuat fungsi dengan dua parameter yaitu img dan level.
di dalama fungsi itu terdapat:

insialisasi hist yang mana berisi indeks level dengan tipe int dan menampilkan nilai dari variabel tinggi dan lebar. lalu terdapat perulangan untuk mengunjungi setiap pixel pada citra. lalu membuat variabel goals yang kemudian akan menyimpan berisikan nilai indeks level
yang di set sebagai nilai integer. lalu mengeksekusi formula dari variabel goal_value dan terakhir menggunakan fungsi return untuk mengembalikan nilai dari gaol_value

```python
target1 = find_goal(gray, 256)
target2 = find_goal(gray_haechan, 256)
```

Code di atas digunakan untuk:

pendeklarasian variabel target1 dan target2 untuk menyimpan nilai find_goal yang memiliki nilai dari gray dan gray_haechan


```python
def find_closest_color(nilai, tujuan):
    min = abs(nilai-tujuan[0])#mendefinisikan variabel color_min bernilai sama dengan nol.
    color_min = 0
    
    for i in range(len(tujuan)):
        cheking = abs(nilai-tujuan[i])
        if(min > cheking):
            min = cheking
            color_min = i
    return color_min 
```

Code di atas digunakan untuk:

pada rumus spesifikasi kita diharuskan untuk mencari nilai terdekat untuk mencapai hasil spesifikasinya dan code di atas merupakan fungsi untuk melakukakn hal tersebut. hal ini dilakukan dengan cara membaut sebuah perulangan dan di dalam perulangan tersebut akan ada sebuag variabel cheking yang akan di masukkan nilai abs(nilai-tujuan[i]). setelah itu akan dibuat kondisi untuk membandingkan variabel min dan cheking. jika min lebih besar dari cheking maka variabel color_min akan diisi oleh nilai perulangan pada saat itu dan pada setiap perulangan tersebut ada fungsi return yang akan mengembalikan nilai dari variabel color_min.

```python
map = np.zeros(256, dtype=int) #membuat 1 array kosong sebanyak 256 yang di set sebagai nilai data integer
for i in range(256): #membuat perulangan untuk mengakses 256 nilai awal dari data
    map[i] = find_closest_color(target1[i], target2)
```

code di atas digunakan untuk:

mendeklarasikan variabel array map yang akan diisi oleh nilai dari proses pemanggilan fungsi find_closest_color dengan parameter 2 variabel yaitu target1 dan target2

```python
spesifikasi = np.zeros(gray.shape).astype(int)
for i in range(gray.shape[0]):
    for j in range(gray.shape[1]):
        spesifikasi[i, j] = map[gray[i, j]]
        #berfungsi untuk menampilkan perbandingan dari histogram img1 dan spesifikasi
plt.title("Citra Spesifikasi Berdasarkan Citra 2")
plt.imshow(spesifikasi, cmap='gray')
```

code di atas digunakan untuk:

mendeklarasikan variabel array spesifikasi yang akan diisi oleh nilai dari proses perulangan yang diisi setiap pixelnhya oleh isi dari variabel map.

Lalu citra hasil spesifikasi akan ditampilkan menggunakan fungsi plt.imshow().

```python
fig, axs = plt.subplots(1, 3, figsize=(12,5))
axs[0].imshow(gray, cmap='gray')
axs[0].set_title('Citra 1')
axs[1].imshow(gray_haechan, cmap='gray')
axs[1].set_title('Citra 2')
axs[2].imshow(spesifikasi, cmap='gray')
axs[2].set_title('Spesifikasi Citra 1')
plt.show()
```
