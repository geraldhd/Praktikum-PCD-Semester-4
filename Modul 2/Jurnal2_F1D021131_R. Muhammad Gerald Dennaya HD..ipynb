Translasi sejauh x = -50 dan y = 40

```python
#translasi sejauh x = -50 dan y = 40
x=-50
y=40
translasi = np.zeros((panjang, lebar)).astype(int)
#===================================================
for i in range(panjang):
    for j in range(lebar):
        if i-40>0 and j+50<lebar :
            translasi[i][j] = gray_img[i-40,j+50]
plt.imshow(translasi, cmap='gray')
```

Rotasi dengan sudut 70 derajat

```python
#rotasi dengan sudut 70 derajat
imgRotate = np.zeros((panjang, lebar)).astype(int)
sudut = np.deg2rad(70)
midRow = int((panjang)/2)
midCol = int((lebar)/2)
#=====================================================
for i in range(panjang):
    for j in range(lebar):
        y = (i - midCol)*np.cos(sudut)+(j-midRow)*np.sin(sudut)
        x = -(i - midCol)*np.sin(sudut)+(j-midRow)*np.cos(sudut)
        
        y = int(y + midCol)
        x = int(x + midRow)
        
        if x>=0 and y>=0 and x<panjang and y<lebar :
            imgRotate[i][j]= gray_img[y][x]
            
            
plt.imshow(imgRotate,cmap="gray")
```

Code di atas digunakan untuk rotate gambar.

Seperti sebelumnya kita perlu menyiapkan variabel-variabel yang akan digunakan seperti imgrotate sebegai variabel utama, sudut yang menggunakan fungsi deg2rad untuk mengubah derajat menjadi radian dan vaeiabel lainnya.

Pencerminan terhadap sumbu x

```python
#pencerminan terhadap sumbu x
cerminx = np.zeros(gray_img.shape)
mirror=panjang-1
for i in range(panjang):
    for j in range(lebar):
        cerminx[i][j]=gray_img[mirror][j]
    mirror-=1 
plt.imshow(cerminx, cmap="gray")
```

Dilatasi matriks dengan faktor skala = 3

```python
#dilatasi matriks dengan faktor skala = 3
#zoom in
dilatation = np.zeros([1305, 1305]) 
for i in range(len(gray_img)):
    for j in range(len(gray_img)): 
        for x in range(i*3, (i*3)+3): 
            for y in range(j*3, (j*3)+3): 
                dilatation[x][y] = gray_img[i][j] 
plt.imshow(dilatation, cmap="gray")
```

```python
plt.imshow(gray_img, cmap="gray")
```

code di atas digunakan untuk menampilkan gambar aslinya

```python
negatif=255-gray_img

#syintax dibawah merupakan sebuah pengaturan untuk menampilakn gambar yang diman subplot untuk mengatur posisi citra yang akan ditampilkan
plt.subplot(121)
plt.imshow(gray_img, cmap='gray')
plt.subplot(122)
plt.imshow(negatif, cmap='gray')

plt.show()
```

Code di atas digunakan untuk enampilkan gambar negatif.

untuk menampilkan gambar negatif hanya perlu menambahkan rumus 255 - gammbar yang ingin dinegatifkan seperti pada code di atas.
lalu penggunaan fungsi subplot digunakan untuk mengatur tampilan sehingga gambar sebelum dan sesudah dinegatifkan dapat disandingkan di kiri dan kanan untuk mempermudah pembandingan. 

```python
grayscale= gray_img
```

code diatas untuk mengembalikan gambar yang sudah dinegatifkan sebelumnya menjadi normal kembali dengan warna abu

```python
derajat = np.zeros(256)

for i in range(256):
    derajat[i]=i
```

melakukan pengulangan untuk membuat variabel array sebagai ukuran acuan dengan range 256

```python
in_gray=np.zeros(256)
for row in grayscale:
    for col in row: 
        in_gray[col]+=1
```

melakukan pengulangan untuk menambahkan nilai 1 pada masing masing pixel pada sebuah citra, gunannya agar dapat merubah citra bernilai 2 dimensi menjadi 1 dimensi

```python
in_negatif=np.zeros(256) 
for row in negatif: 
    for col in row: 
        in_negatif[col]+=1
```

melakukan pengulangan untuk menambahkan nilai 1 pada masing masing pixel pada sebuah citra, gunannya agar dapat merubah citra bernilai 2 dimensi menjadi 1 dimensi

```python
plt.subplot(121)
plt.xlabel("keabuan")
plt.ylabel("jumlah pixel")
plt.bar(derajat,in_gray) 
plt.subplot(122)
plt.xlabel("negatif") 
plt.ylabel("jumlah pixel")
plt.bar(derajat,in_negatif) 
```

Code di atas digunakan untuk menampilkan histogram dari gambar sebelum dan sesudah negatif.

untuk menampilkan gambar negatif hanya perlu menambahkan rumus 255 - gammbar yang ingin dinegatifkan seperti pada code di atas.
