A. Translasi sejauh x = -50 dan y = 40

```python
img = cv2.imread('beanny.jpg',0)
a,b = img.shape
 
M = np.float32([[1,0,-50],[0,1,40]])
dst = cv2.warpAffine(img,M,(a,b))
 
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows
```

B. Rotasi dengan sudut 70 derajat

```python
rotate = cv2.imread('beanny.jpg', 1)
i, j = rotate.shape[1], rotate.shape[0]

rotation = cv2.getRotationMatrix2D((i/2, j/2), 70, 1)
img_rotate = cv2.warpAffine(rotate, rotation, (i,j))

cv2.imshow('Image Rotate',img_rotate)
cv2.waitKey(0)
cv2.destroyAllWindows
```

D. Dilatasi matriks dengan faktor skala = 3

```python

img = cv2.imread('beanny.jpg', cv2.IMREAD_UNCHANGED)
 
print('Original Dimensions : ',img.shape)
 
scale_percent = 300 # percent of original size
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)
 
# resize image
resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
 
print('Resized Dimensions : ',resized.shape)
 
cv2.imshow("Resized image", resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Soal 2

Lakukan Operasi Citra Negatif dari citra grayscale lalu bandingkan Histogram kedua citra tersebut.

```python
negatif=255-gray_img

#syintax dibawah merupakan sebuah pengaturan untuk menampilakn gambar yang diman subplot untuk mengatur posisi citra yang akan ditampilkan
plt.subplot(121)
plt.imshow(gray_img, cmap='gray')
plt.subplot(122)
plt.imshow(negatif, cmap='gray')

plt.show()
```

Pembandingan Histogram

```python
grayscale= gray_img

derajat = np.zeros(256)

for i in range(256):
    derajat[i]=i

in_gray=np.zeros(256)
for row in grayscale:
    for col in row: 
        in_gray[col]+=1

in_negatif=np.zeros(256) 
for row in negatif: 
    for col in row: 
        in_negatif[col]+=1
plt.subplot(121)
plt.xlabel("keabuan")
plt.ylabel("jumlah pixel")
plt.bar(derajat,in_gray) 
plt.subplot(122)
plt.xlabel("negatif") 
plt.ylabel("jumlah pixel")
plt.bar(derajat,in_negatif) 
```
